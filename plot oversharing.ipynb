```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

path_spc = "C:\\Users\\root\\spc_optimized_reps.csv"
path_nn = "C:\\Users\\root\\nn_reps.csv"
```

```python
res_nn = pd.read_csv(path_nn)
res_spc = pd.read_csv(path_spc)
spcs_nn = res_nn.groupby("specialization power").mean()["specialization"].values
spcs_err_nn = res_nn.groupby("specialization power").std()['specialization'].values
alphas = res_nn.groupby("specialization power").mean()["specialization"].index
spcs_spc = res_spc.groupby("specialization power").mean()["specialization"].values
spcs_err_spc = res_spc.groupby("specialization power").std()['specialization'].values
plt.figure(figsize=(8,6))
plt.errorbar(alphas,y=spcs_nn,yerr=spcs_err_nn,label="NN",marker='o',markersize=15)
plt.errorbar(alphas,y=spcs_spc,yerr=spcs_err_spc,label="SPC",marker='o',markersize=15)
plt.xlabel("alpha")
plt.ylabel("specialization")
plt.legend(loc=2)
plt.title("Evolving interaction strength 10 reps, 10000 steps, pop=1000");
```
